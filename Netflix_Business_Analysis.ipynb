```python
df = df.where(pd.notnull(df),None)
```

```python
df.head()
```

```python
unique_genre = df['listed_in'].str.split(', ').explode().dropna().unique()
print(unique_genre)
print("Count_unique_gener:",len(unique_genre))
```

```python
unique_country = df['country'].str.split(', ').explode().dropna().unique()
print(unique_country)
print("Total_unique_country:",len(unique_country))
```

```python
unique_rating = df['rating'].dropna().unique()
print(unique_rating)
print("Total_unique_rating:",len(unique_rating))
```

```python
years = df['release_year'].dropna().unique()
print(years)
print("Total_year_count:",len(years))
```

```python
df = df.drop(columns = ['description'])
```
